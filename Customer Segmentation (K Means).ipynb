```python
x1=df.loc[:,['Age','Spending Score (1-100)']].values
# extracts all the values of the 'Age' and 'Spending Score (1-100)' columns from the DataFrame df

from sklearn.cluster import KMeans

WCSS=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(x1)
    WCSS.append(kmeans.inertia_)
plt.figure(figsize=(10,6))
plt.grid()
plt.plot(range(1,11),WCSS,color="red")
plt.xlabel("K-value")
plt.ylabel("WCSS")
plt.show()
```

```python
#shows whic data point belong to which cluster
kmeans=KMeans(n_clusters=4)
belonging=kmeans.fit_predict(x1)
print(belonging)
```

```python
kmeans.cluster_centers_
```

```python
sns.scatterplot(x=x1[:,0],y=x1[:,1],c=kmeans.labels_, cmap='rainbow')
sns.scatterplot(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1],color='black')
plt.title("Customers' clusters")
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.grid()
plt.show()
```
